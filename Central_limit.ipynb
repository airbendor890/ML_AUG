### Central Limit theorem Visulization

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
seq=[1,1,2,2,2,2,2,2,2,3,3,5,6,6,6,6,6,6,6,6,6,6,5,5,5,4,4,4]
N=30
M=1000
L1=[[random.choice(seq) for i in range(N)] for j in range(M)]
S=[sum(L1[j]) for j in range(M)]
```

```python
fig=plt.figure(figsize=(10,8))
plt.hist(S,bins=50,histtype='step',density=True);
```

```python
E=0
for i in range(1,7):
    E+=i*seq.count(i)/len(seq)
print(E*N)
```

```python
Lambda=0.2
N=30
M=1000
L1=[[random.expovariate(Lambda) for i in range(N)] for j in range(M)]
S=[sum(L1[j]) for j in range(M)]
fig=plt.figure(figsize=(10,8))
plt.hist(S,bins=50,histtype='step',density=True);
```

```python
print(Lambda**-1*N)
```

```python
N=30
M=1000
L1=[[random.gauss(5,1) for i in range(N)] for j in range(M)]
S=[sum(L1[j]) for j in range(M)]
fig=plt.figure(figsize=(10,8))
plt.hist(S,bins=50,histtype='step',density=True);
```
